```
class CurveDataset(Dataset):
    def __init__(self, img_dir, mask_dir=None, transform=None):
        self.imgs = sorted(glob.glob(os.path.join(img_dir, '*.png')))
        self.transform = transform

    def __len__(self):
        return len(self.imgs)

    def __getitem__(self, idx):
        img_path = self.imgs[idx]
        img = Image.open(img_path).convert('RGB')
        img = np.array(img)

        if self.transform:
            img = self.transform(image=img)['image']
        return img
```

```
Ï = None

def preprocess_edges_dark(img_rgb, low_thresh=0.1, high_thresh=0.6):
    #denoised = cv2.fastNlMeansDenoisingColored(img_rgb, None, 10, 10, 7, 21)
    #gray = cv2.cvtColor(denoised, cv2.COLOR_RGB2GRAY)

    gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
    blurred = cv2.GaussianBlur(gray, (5,5), 1.41)
    edges = cv2.Canny(blurred, int(low_thresh*255), int(high_thresh*255), apertureSize=3, L2gradient=True)
    edges  = cv2.dilate(edges, (3, 3), iterations=1)
    return edges

def contours_from_mask(mask, min_len=50, min_span_ratio=0.1):
    h, w = mask.shape
    cnts,_ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    good=[]
    for c in cnts:
        if len(c) < min_len:
          continue
        if np.ptp(c[:,0,0]) / w < min_span_ratio:
          continue
        good.append(c[:,0,:])
    return good

def fit_sinusoid(points, Ï, h, guess=None, max_pts=2000):
    points = np.asarray(points)
    xs, ys = points[:, 0], points[:, 1]

    if len(xs) > max_pts:
        sel = np.random.choice(len(xs), max_pts, replace=False)
        xs, ys = xs[sel], ys[sel]

    order = np.argsort(xs)
    xs, ys = xs[order], ys[order]

    ys = median_filter(ys, size=5)

    if guess is None:
        guess = (np.ptp(ys)/2, 0.0, ys.mean())

    def model(x, A, Î¸, y0):
        return A * np.sin(Ï * x + Î¸) + y0

    try:
        A, Î¸, y0 = curve_fit(model, xs, ys, p0=guess, maxfev=5000)[0]
        if A < 0:
            A = -A
            Î¸ = (Î¸ + math.pi) % (2 * math.pi)
        return A, Î¸, y0
    except RuntimeError:
        return None

def split_upper_lower(pts: np.ndarray, min_x_span: int = 50):
    xs, ys = pts[:, 0], pts[:, 1]
    uniq_x = np.unique(xs)
    if np.ptp(uniq_x) < min_x_span:
        return []

    upper, lower = [], []
    for x in uniq_x:
        ys_here = ys[xs == x]
        upper.append((x, ys_here.min()))
        lower.append((x, ys_here.max()))
    return [np.array(upper), np.array(lower)]

def detect_and_draw(img_rgb,
                    background="rgb",
                    top_N=3,
                    rms_thr=10.0,
                    min_len=40,
                    span_ratio=0.03,
                    show_debug=False):

    h, w = img_rgb.shape[:2]
    Ï = 2 * np.pi / w

    edges = preprocess_edges_dark(img_rgb)

    contours = contours_from_mask(edges, min_len, span_ratio)

    #grad_mag = cv2.Sobel(edges, cv2.CV_32F, 1, 0, ksize=3)**2 + cv2.Sobel(edges, cv2.CV_32F, 0, 1, ksize=3)**2
    #grad_ang = np.rad2deg(np.arctan2( cv2.Sobel(edges, cv2.CV_32F, 0, 1, ksize=3), cv2.Sobel(edges, cv2.CV_32F, 1, 0, ksize=3)))
    #mask_horiz = (np.abs(grad_ang) < 45).astype(np.uint8)
    #edges = cv2.bitwise_and(edges, edges, mask=mask_horiz)
    #print (edges)

    scored = []

    for cnt in contours:
        for pts in split_upper_lower(cnt, min_x_span=30):
            if len(pts) < 30:
                continue

            fit = fit_sinusoid(pts, Ï, h)
            if fit is None:
                continue
            A, Î¸, y0 = fit

            if A < 2 or A > h / 1.5:
                continue

            xs, ys = pts[:, 0], pts[:, 1]
            rmse = np.sqrt(((ys - (A*np.sin(Ï*xs + Î¸) + y0)) ** 2).mean())
            if rmse > 10:
                continue

            geom = len(pts) / rmse
            if A > 200:
                continue
            scored.append((geom, A, Î¸, y0))

    scored.sort(key=lambda t: t[0], reverse=True)
    curves = [(A, Î¸, y0) for _, A, Î¸, y0 in scored[:top_N]]

    canvas = cv2.cvtColor(edges, cv2.COLOR_GRAY2BGR) if background == "mask" else img_rgb.copy()

    xs_full = np.arange(w)
    for A, Î¸, y0 in curves:
        ys = (A * np.sin(Ï*xs_full + Î¸) + y0).astype(int)
        good = (ys >= 0) & (ys < h)
        pts = np.column_stack([xs_full[good], ys[good]]).astype(np.int32)
        cv2.polylines(canvas, [pts], False, (0, 255, 0), 3)

    plt.figure(figsize=(6, 6))
    plt.imshow(canvas)
    plt.axis("off")
    plt.title(f"{len(curves)} ÑÑÑÐºÐ¸")
    plt.show()

    for c in contours:
        plt.plot(c[:, 0], c[:, 1], '.', markersize=1)
    plt.gca().invert_yaxis()
    plt.title("ÐÐ¾Ð½ÑÑÑÑ Ð´Ð¾ fit_sin")
    plt.show()

    if show_debug:
        plt.figure(figsize=(4, 4))
        plt.imshow(edges, cmap="gray")
        plt.axis("off")
        plt.title("edges")
        plt.show()

for i in range (49, 59 + 1):
    sample_path = ds.imgs[i]
    img = np.array(Image.open(sample_path).convert('RGB'))
    detect_and_draw(img_rgb=img, show_debug=True)

sample_path = ds.imgs[100]
img = np.array(Image.open(sample_path).convert('RGB'))
detect_and_draw(img_rgb=img, show_debug=True)
```

```
def draw_sine(img, A, Î¸, y0, Ï, color=(0, 255, 0), thickness=3):
    h, w = img.shape[:2]
    xs = np.arange(w)
    ys = (A * np.sin(Ï * xs + Î¸) + y0).astype(int)
    mask = (ys >= 0) & (ys < h)
    pts = np.column_stack([xs[mask], ys[mask]]).astype(np.int32)
    if len(pts) > 1:
        cv2.polylines(img, [pts], isClosed=False, color=color, thickness=thickness)
```

```
def clean_rows_by_depth_from_reference(rows_4, rows_1, threshold=0.1):
    rows_4 = sorted(rows_4, key=lambda r: r[0])
    rows_1 = sorted(rows_1, key=lambda r: r[0])

    keep_indices = set()
    removed_indices = set()

    for r1 in rows_1:
        d1 = r1[0]
        close_indices = [i for i, r4 in enumerate(rows_4) if abs(r4[0] - d1) < threshold]

        if close_indices:
            closest_idx = min(close_indices, key=lambda i: abs(rows_4[i][0] - d1))
            keep_indices.add(closest_idx)
            for i in close_indices:
                if i != closest_idx:
                    removed_indices.add(i)

    result = [r for i, r in enumerate(rows_4) if i in keep_indices or (i not in removed_indices and all(abs(r[0] - r1[0]) >= threshold for r1 in rows_1))]

    return sorted(result, key=lambda r: r[0])

unique_rows_4 = unique_rows
unique_rows_1 = pd.read_csv("/content/sinusoids_colored_mpt1.csv").values.tolist()
unique_rows = clean_rows_by_depth_from_reference(unique_rows_4, unique_rows_1)
```
